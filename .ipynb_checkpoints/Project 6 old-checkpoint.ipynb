# Reduce Demensionality and Scale Features

```python
steam_df_new = steam_df.copy()
#appid isn't a meaningful statistic
steam_df_new.drop("appid", axis=1, inplace=True)
#owners is too vague to be very useful
steam_df_new.drop("owners", axis=1, inplace=True)
#english isn't explained, honestly don't know exactly what it means
steam_df_new.drop("english", axis=1, inplace=True)

steam_df_new_Y = steam_df_new[['positive_ratings', 'negative_ratings']]

steam_df_Y = steam_df_new_Y['positive_ratings']/(steam_df_new_Y['negative_ratings']+steam_df_new_Y['positive_ratings'])

steam_df_new.drop('positive_ratings', axis=1, inplace=True)
steam_df_new.drop('negative_ratings', axis=1, inplace=True)

scaled_features = steam_df_new.copy()
scaled_features['release_date'] = pd.to_datetime(scaled_features['release_date']).astype('int64')

tempList = []
for plat in scaled_features['platforms']:
    tempPlat = plat.split(';')
    tempList.append(tempPlat)
scaled_features['platforms'] = tempList

tempList = []
for cat in scaled_features['categories']:
    tempCat = cat.split(';')
    tempList.append(tempCat)
scaled_features['categories'] = tempList

tempList = []
for genre in scaled_features['genres']:
    tempGenre = genre.split(';')
    tempList.append(tempGenre)
scaled_features['genres'] = tempList

tempList = []
for tag in scaled_features['steamspy_tags']:
    tempTag = tag.split(';')
    tempList.append(tempTag)
scaled_features['steamspy_tags'] = tempList

#names of columns to be scaled
col_names = ['release_date', 'required_age', 'achievements', 'average_playtime', 'median_playtime', 'price']
features = scaled_features[col_names]
scaler = preprocessing.StandardScaler().fit(features.values)
features = scaler.transform(features.values)

scaled_features[col_names] = features
# scaled_features.drop("appid", axis=1, inplace=True)
scaled_features

# x = steam_df_new.values #returns a numpy array
# scaler = preprocessing.StandardScaler()
# x_scaled = scaler.fit_transform(x)
# steam_df_new = pd.DataFrame(x_scaled, columns = steam_df_new.columns)

# scaler = StandardScaler()
# scaler.fit(steam_df_new["achievements"])
# steam_df_new["achievements"] = scaler.transform(steam_df_new["achievements"])
```
